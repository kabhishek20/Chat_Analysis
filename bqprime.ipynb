```python
import openai
import requests
import pandas as pd
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
```

```python
url = "https://news.google.com/rss/search?q=FDA%20483%20site%3Awww.bqprime.com%20when%3A1y&hl=en-IN&gl=IN&ceid=IN%3Aen"
# Download the XML data
response = requests.get(url)
xml_data = response.content

# Parse the XML data
root = ET.fromstring(xml_data)

# Extract information from the XML tree and create a dictionary
data = []
for item in root.iter('item'):
    title = item.find('title').text
    link = item.find('link').text

    data.append({'Title': title, 'URL': link})

# Create a DataFrame from the extracted data
df = pd.DataFrame(data)
df
```

```python
def get_date(soup):
    story_elements = soup.find_all(class_="story-base-template-m__story-date__3YCRm")
    content = [element.get_text() for element in story_elements]
    return content
```

```python
def get_soup(url):
  response = requests.get(url)
  if response.status_code == 200:
    page_content = response.text
    soup = BeautifulSoup(page_content, 'html.parser')
    return soup
  else:
    print("Failed to retrieve the page. Status code:", response.status_code)
    return None
```

```python
raw_data=[]
date = []
for index, row in df.iterrows():
    # Access the values of each column for the current row
    url = row['URL']
    soup=get_soup(url)
    if soup:
      content = get_content(soup)
      time_stamp = get_date(soup)
      raw_data.append(" ".join(content[1:]))
      date.append(time_stamp)
    else:
      pass
```

```python
df["raw_data"]=raw_data
df["date"]=date
df
```
