```python
from bs4 import BeautifulSoup
import requests
import xml.etree.ElementTree as ET
import pandas as pd
import openai
import datefinder
```

```python

url = "https://news.google.com/rss/search?q=FDA%20483%20site%3Amedicaldialogues.in%20when%3A1y&hl=en-IN&gl=IN&ceid=IN%3Aen"

# Download the XML data
response = requests.get(url)
xml_data = response.content

# Parse the XML data
root = ET.fromstring(xml_data)

# Extract information from the XML tree and create a dictionary
data = []
for item in root.iter('item'):
    title = item.find('title').text
    link = item.find('link').text

    data.append({'Title': title, 'URL': link})

# Create a DataFrame from the extracted data
df = pd.DataFrame(data)
df
```

```python
df["URL"][0]
```

```python
def get_content(soup):
    story_elements = soup.find_all(class_="story")
    story_elements_1 = soup.find_all(class_="pasted-from-word-wrapper")
    content = [element.get_text() for element in story_elements]
    content1 = [element.get_text() for element in story_elements_1]
    content.extend(content1)
    return content
```

```python
def get_date(soup):
    story_elements = soup.find_all(class_="post-published updated")
    content = [element.get_text() for element in story_elements]
    return content
```

```python
def get_soup(url):
  response = requests.get(url)
  if response.status_code == 200:
    page_content = response.text
    soup = BeautifulSoup(page_content, 'html.parser')
    return soup
  else:
    print("Failed to retrieve the page. Status code:", response.status_code)
    return None
```

```python
raw_data=[]
date = []
for index, row in df.iterrows():
    # Access the values of each column for the current row
    url = row['URL']
    soup=get_soup(url)
    # print(soup)
    if soup:
      content = get_content(soup)
      time_stamp = get_date(soup)
      if len(time_stamp)==0:
        time_stamp = "None"
      raw_data.append(" ".join(content))
      date.append(time_stamp)
    else:
      pass
```

```python
df["raw_data"]=raw_data
df["date"]=date
df
```

```python
df = df[df["raw_data"]!=""]
```
